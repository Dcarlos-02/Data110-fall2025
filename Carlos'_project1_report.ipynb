--importing libraries

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

plt.style.use('default')


--loading dataset

df = pd.read_csv("https://raw.githubusercontent.com/Dcarlos-02/Data110-fall2025/main/infantmortality.csv")

df.head(20)

--getting information about the dataset such as null value counts and data types

df.info()

--renaming the columns for better readability

df.rename(columns={'Maternal Race or Ethnicity':'ethnicity'}, inplace=True)

df.head()

--Normalizing the data by dropping missing values

df = df.dropna()

df.reset_index(inplace=True, drop=True)

--adjusting data for better

df['ethnicity'] = df['ethnicity'].replace('Asian and Pacific Islander', 'Asian')

df['ethnicity'] = df['ethnicity'].replace('Non-Hispanic Black', 'Black')

df['ethnicity'] = df['ethnicity'].replace('Non-Hispanic White', 'White')

df['ethnicity'] = df['ethnicity'].replace(['Other Hispanic', 'Puerto Rican'], 'Hispanic')

df['ethnicity'] = df['ethnicity'].replace('White Non-Hispanic', 'White')

df['ethnicity'] = df['ethnicity'].replace('Black Non-Hispanic', 'Black')

df.head(20)

--getting a summary of the data

df.describe()

--creating a line plot to visualize changes over time


df = df.sort_values("Year")

plt.figure(figsize=(10,6))

sns.lineplot(x='Year', y='Infant Mortality Rate', hue=df['ethnicity'], data=df, ci=None)

plt.title(" Infant mortality rate from 2007-2016 by race ")

plt.show()

--creating a boxplot to compare the distribution accross races

plt.figure(figsize=(10,6))

sns.boxplot(x='ethnicity', y='Infant Deaths', data=df)

plt.title("Infant deaths distribution by race")

--creating a scatterplot to correlate live birth count with infant mortality rate

plt.figure(figsize=(10,6))

plt.scatter(df['Number of Live Births'], df['Infant Mortality Rate'], color='orange')

plt.xlabel('Number of live births')

plt.ylabel('kid''s death')

plt.title("Live birth vs Infant deaths")
