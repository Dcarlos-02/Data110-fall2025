#**Air quality: Assessing the implication of meteorological and chemical factors**

#Introduction

Air quality is an impactful factor in the environment. It affects human health, the ecosystem, and overall well-being. Polluted air is linked to several health complications from cardiovascular issues to respiratory issues, making monitoring it a critical task.

In consequence, This project focuses on analyzing air quality data collected in the District of Columbia. It aims to uncover patterns, trends, and relationships among the variables using tools like plotly, pandas, and more... through python. It uses a dataset provided by Opendata, accessible through "https://opendata.dc.gov/datasets/DCGIS::air-quality-purple-air-sensors/about"

The project is led by the following question: How does VOC concentration variate , and how does atmospheric pressure influence concentrations of  particulate matter (PM1, PM2.5, PM10) in the study area?

The dataset contains 185,747 observations distributed across 16 variables. Key variables include:

- VOC: volatile organic compounds
- Humidity: percentage of humidity
- pressure: air pressure
- temperature: current temperature
- latitude: geographical coordinate
- longitude ; geographical coordinate
- pm1: particulate matter of 1 micrometer or smaller
- pm10: particulate matter with diameter of 10 micrometers or less
- pm2.5: particulate matter with diameter of 2.5 or less

#Understanding the relationship between the variables

Below is a heatmap representing the correlation between the variables. Variables such as PM1, PM2.5, and PM10 show a strong positive correlation, indicating that increases in the concentration of particulate matter smaller than 1 micrometer are often associated with increases in particulate matter of 2.5 or 10 micrometers, and vice versa. In contrast, the remaining variables exhibit weak correlations, suggesting limited linear relationships and minimal direct influence on one another. This strong association among particulate matter measures is expected, as they often originate from similar emission sources and are influenced by the same atmospheric conditions. Consequently, changes in air quality events are likely to affect multiple particulate size categories simultaneously.

#Average Daily air pressure

This line plot illustrates the daily air pressure over the study period. Overall, air pressure fluctuates across the time period, indicating frequent changes rather than a stable trend. These variations may reflect shifting weather patterns and atmospheric conditions throughout the observation window. Such fluctuations can influence the dispersion and accumulation of air pollutants, including particulate matter and VOCs. Periods of higher pressure are often associated with more stagnant air conditions, which can allow pollutants to build up near the surface. Conversely, lower pressure systems may enhance air movement and contribute to improved air quality.
