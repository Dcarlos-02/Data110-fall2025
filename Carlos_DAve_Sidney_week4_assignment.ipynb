Week 4 assignment: Bar graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ["aggravated assault", "Burglary", "Larceny", "Motor vehicle theft", "Murder", "Rape", "Robbery"]
y = [264.1, 250.7, 1324.2, 318.7, 5.7, 38.0, 66.5]
z = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
```

 **Task1**

```python
plt.bar(x,y, color='red')
plt.title("Crime rates in 2023")
plt.xlabel('Offense')
plt.ylabel("Offense per 100,00 people")
plt.legend()
plt.grid()
plt.ylim(0,1400)
plt.show
```

**Task 2**

```python
colors=['blue','yellow','green','red','purple','orange','black']
plt.barh(x,z, color=colors )
plt.title("Percentage change in offeneses rates")
plt.xlabel("Change in offense rates")
plt.ylabel("offenses")


```

```python
colors=['blue','yellow','green','red','purple','orange','turquoise']
plt.style.use('dark_background')
plt.barh(x,z, color= ['r' if v<0 else 'blue' for v in z ])
plt.axvline(0,linewidth=3)
plt.grid(False)
plt.title("Percentage change in offeneses rates")
plt.xlabel("Change in offense rates")
plt.ylabel("offenses")
```

The second bar graph makes more sense than the first one because it shows how crime rates have changed over the course of the year. The first one shows the quantity of crime commited, but the second one shows if the rate of criminaity rises or decreases. Therefore, there will be a better understanding of the effectiveness of the decision made,and now the authorities will know where to concentrate more effort.

```python
companies = ["Company A", "Company B", "Company C"]
plt.style.use('default')
product1 = [20, 34, 30]
product2 = [25, 32, 34]
product3 = [30, 35, 27]
x_indexes=np.arange(len(companies))
plt.bar(x_indexes-0.135,product1,width=0.15,label='product1' )
plt.bar(x_indexes,product2,width=0.15,label='product2' )
plt.bar(x_indexes+0.135,product3,width=0.15,label='product3' )
plt.xticks(ticks=x_indexes, labels=companies)
plt.legend()
plt.title("Products' sales value")
plt.ylabel("products")
plt.show()
```
