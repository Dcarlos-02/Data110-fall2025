# **Titanic distributions**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

This dataset contains 891 observations spread accross 15 variables. However, only the following ones will be used: survived, pclass, sex, age, fare, embark_town, sibsp, parch. These variables will help the analysis because they provide key information that can shape death and survival rate.

# Task **2**

```python
plt.hist(titanic['fare'], edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title("Fare distribution")
plt.show()
```

```python
sns.kdeplot(titanic['fare'], fill=True)
plt.title('Density estimate of fare for the Titanic passengers')
plt.show()
```

The distribution is skewed right which implies that the distibution is more elevated accross the low values. However, There is a few observations above 500, which we can consider as outliers. As a result, we can conclude that we have a wide majority of the passengers in the cheapest class.

# Task **3**

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x='fare', hue='survived', fill=True, alpha=0.1)
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

More people survived in the cheapest class--which make sense because their distribution was higher. Approximately the same amount of people died and survived in what can call the middle classes. However, nobody survived in the higher classes.

# Task **4**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='pclass', y='fare', hue='sex', data=titanic)
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.title("Fare distribution by passenger class and sex")
plt.show()
```

In the first class, the median fare is higher for woman. However, the distribution does not vary significantly accross the classes.

# Task **5**

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='sex', y='age', data=titanic, color='r')
plt.title("Distribution of age by sex")
plt.show()
```

The median age of passengers for both sex is approximately 30 years old.

# Task **6**

```python
sns.stripplot(x='pclass', y='fare', data=titanic, jitter=0.35)
plt.title('')
plt.show()
```
